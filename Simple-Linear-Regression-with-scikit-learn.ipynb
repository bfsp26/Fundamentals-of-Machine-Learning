# **Simple Linear Regression with Scikit-Learn**

```python
import pandas as pd
url = 'https://raw.githubusercontent.com/jeffprosise/Machine-Learning/master/Data/poverty.csv' # poverty.csv
df = pd.read_csv(url)
df.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

x = df['PovPct'].values.reshape(-1, 1)
y = df['TeenBrth'].values.reshape(-1, 1)
plt.scatter(x, y)
plt.xlabel('Percentage of Residents Living Below the Poverty Line')
plt.ylabel('Teen Birth Rate')
plt.title('Teen Birth Rate as a Function of Poverty Level')
plt.show();
```

```python
from sklearn.linear_model import LinearRegression

LR = LinearRegression()
LR.fit(x, y)
```

```python
plt.scatter(x, y)
plt.plot(x, LR.predict(x), color='r')
plt.xlabel('Percentage of Residents Living Below the Poverty Line')
plt.ylabel('Teen Birth Rate')
plt.title('Teen Birth Rate as a Function of Poverty Level')
plt.show();
```
